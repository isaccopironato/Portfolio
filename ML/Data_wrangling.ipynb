```python
from ucimlrepo import fetch_ucirepo
import pandas as pd
import seaborn as sns
import numpy as ny
"""
# fetch dataset
online_retail = fetch_ucirepo(id=352)

# data (as pandas dataframes)
X = online_retail.data.features
y = online_retail.data.targets

# metadata
print(online_retail.metadata)

# variable information
print(online_retail.variables)
"""

df = pd.read_csv('https://archive.ics.uci.edu/static/public/352/data.csv')
```

```python
df.head(50)
```

```python
pd.to_datetime(df['InvoiceDate'])
```

```python
df["anno"] = pd.DatetimeIndex(df['InvoiceDate']).year
df["mese"] = pd.DatetimeIndex(df['InvoiceDate']).month
df
```

```python
df['Quarter'] = pd.DatetimeIndex(df['InvoiceDate']).year.astype(str) + '-Q' + pd.DatetimeIndex(df['InvoiceDate']).quarter.astype(str)
df
```

```python
df.describe()
```

```python
df_country = df.groupby("Country").size().reset_index(name='Count')
df_country
```

```python
df_country['Percentage'] = (df_country['Count'] / df_country['Count'].sum() * 100).round(4)
df_country = df_country.sort_values('Percentage', ascending=False)
df_country
```

```python
temp = df.groupby("Country")["UnitPrice"].sum()
df_country["Total_rev"] = df_country["Country"].map(temp)
df_country
```

```python
df_sample_mille = df.sample(n = 1000)
df_sample_mille
```

```python
df_sample_10_percento = df.sample(frac= 0.1)
df_sample_10_percento
```

```python
df_sample_10_percento_con_rimpiazzo = df.sample(frac= 0.1, replace= True)
df_sample_10_percento_con_rimpiazzo
```

```python
df_stratificato_10_percento = df_country.apply(lambda x: x.sample(frac = 0.1))
df_stratificato_10_percento.describe()
```

```python
df_pivot_year = df.pivot_table(index = "Country", columns = "anno" , values = "UnitPrice", aggfunc = "sum")
df_pivot_year
```

```python
df_pivot_quarter = df.pivot_table(index = "Country", columns = "Quarter" , values = "UnitPrice", aggfunc = "sum")
df_pivot_quarter
```

```python
df_pivot_month = df.pivot_table(index = "Country", columns = ["anno","mese"] , values = "UnitPrice", aggfunc = "sum")
df_pivot_month
```
