```python
import pandas as pd
import re

df = pd.read_csv(path + "/spam.csv", encoding="latin-1")
df.head()
```

```python
df["v2"] = df["v2"].str.lower()
df["v2"].head(50)
```

```python
df["v2"] = df["v2"].str.strip()
print(df["v2"][0])
```

```python
t = [" the ", " and ", " or ", "  "]

for i in t:
  df["v2"] = df["v2"].str.replace(i, " ")

```

```python
temp = df["v2"].str.extractall(r"([0-9+ ]{8,15})")
temp.head()
```

```python
df["v2"] = df["v2"].str.replace(r"([0-9+ ]{8,15})", "")
df["v2"].head()
```

```python
df["v2"].str.extractall(r"([a-z0-9.-_%]+@[a-z0-9-_%]+\.{2,})")
```

```python
df["v2"] = df["v2"].str.replace(r"([a-z0-9.-_%]+@[a-z0-9-_%]+\.{2,})", "")
df["v2"].head()
```

```python
df["v2"].str.extractall(r"([a-z0-9-_%]+\.[a-z]{2,3})")
```

```python
df["v2"] = df["v2"].str.replace(r"([a-z0-9-_%]+\.[a-z]{2,3})", "")
df["v2"].head()
```

```python

df["v2"] = df["v2"].str.replace("  ", " ")

```

```python
temp_my = df
temp_my["lista_parole"] = df["v2"].str.split(" ", expand = False)
temp_my.head()
```

```python
words_d = {}
for words in temp_my["lista_parole"]:
  for word in words:
    if word in words_d:
      words_d[word][0] += 1
      words_d[word][1] += len(words)
    else:
      words_d[word] = [1, len(words)]

words_d
```

```python
words_d = dict(sorted(words_d.items(), key=lambda item: item[1], reverse = True))
words_d
```

```python
d_to_df = {
           "words" : words_d.keys(),
           "count" : [value[0] for value in words_d.values()],
           "len_text" : [value[1]/value[0] for value in words_d.values()]
           }

df_count = pd.DataFrame.from_dict(d_to_df)
df_count.head()
```

```python
df_count.sort_values(by = "count", ascending = False, inplace = True)
df_count.head()
```

```python
df_count["len_word"] = df_count["words"].str.len()
df_count.head(50)
```
