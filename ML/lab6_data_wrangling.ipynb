```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('https://archive.ics.uci.edu/static/public/45/data.csv')
df.describe()
```

```python
df.head(20)
```

# Encoding

## discretizazione

Rendere piu discreto le eta in modo da riminuire il range di ita in fasi di vita comunemente acettate

```python
"""
child = 0-15
teen = 15-20
adult = 20-40
middle_aged = 40-65
senior = 65+
"""

df["age_range"] = pd.cut(df["age"], bins=[0, 15, 20, 40, 65, float("inf")], labels=["child", "teen", "adult", "middle_aged", "senior"])
df.head()
```

## One-hot encoding

```python
df_hot_one_encoding = pd.get_dummies(df, columns = ["sex"])
df_hot_one_encoding.head()
```

```python
df_hot_one_encoding_2 = pd.get_dummies(df, columns = ["age_range"], prefix="", prefix_sep="")
df_hot_one_encoding_2.head()
```

## Lable encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lable_encoder = LabelEncoder()
df_label_encoding = df.copy()
df_label_encoding["age_lable"] = lable_encoder.fit_transform(df["age_range"])
df_label_encoding.head(50)
```

## Binary encoding

```python
!pip install category_encoders
```

```python
import category_encoders as ce
```

```python
df_dinary_encoding = df.copy()
binary_encoder = ce.BinaryEncoder()
df_dinary_encoding = binary_encoder.fit_transform(df_dinary_encoding)
df_dinary_encoding.head()
```

## Target encoding


```python
df_target_encoding = df.copy()
means = df_target_encoding.groupby("age_range")["chol"].mean()
df_target_encoding["category_encoded"] = df_target_encoding["age_range"].map(means)
df_target_encoding.head()
```

## Frequency encoding


```python
df_frequency_encoding = df.copy()
freq = df_frequency_encoding['age_range'].value_counts(normalize = True)*100
df_frequency_encoding["frequency"] = df_frequency_encoding["age_range"].map(freq).round(2)
df_frequency_encoding.head()
```

# Normalizazione

## Distribuzioni

```python
df.head()
```

```python
numeric_colums = df.select_dtypes(include=np.number).columns.tolist()
numeric_colums
```

```python
for i in numeric_colums:
    plt.figure(figsize=(8, 6))  # Create new figure for each plot
    sns.histplot(df[i])
    plt.title(f'Distribution of {i}')
    plt.xlabel(i)
    plt.ylabel('Frequency')
    plt.grid(True, alpha=0.3)
    plt.show()
```

## Min-Max scaler

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
df_min_max = df.copy()
min_max_scaler = MinMaxScaler()
df_min_max = pd.DataFrame(min_max_scaler.fit_transform(df_min_max[numeric_colums]), columns = numeric_colums)
```

## L1/L2 Normalization
